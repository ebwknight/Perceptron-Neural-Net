```python
import numpy as np


file = open('digit-examples-all.txt', 'r') 
lines = file.readlines()
print(lines)
# splitlines = lines.splitlines()
# print(splitlines)

    
```

```python
numDataPoints = len(lines)//2

input_data = np.empty(shape=(numDataPoints ,2, 64))

bitMaps = []
numAnswers = []

for i in range(len(lines)):
    if i % 2 == 0:
        bitMaps.append(lines[i].split()[1:-1])
    else:
        numAnswers.append(lines[i].split()[1:-1])
```

```python
print(bitMaps[0])
```

```python
print(numAnswers[0])
singleNumAnswers = []
```

```python
for i in range(len(numAnswers)):
    for j in range(len(numAnswers[i])):
        if numAnswers[i][j] == '1.0':
            singleNumAnswers.append(j)
print(singleNumAnswers[0])
```

```python
print(input_data)
```

```python
for i in range(numDataPoints):
    input_data[i][0] = bitMaps[i]
    input_data[i][1] = singleNumAnswers[i]
```

```python
print(input_data[0])
#the actual answer is just the answer 64 times since each element in the 2d array has 64 slots
```
